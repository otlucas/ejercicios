```python
import pyspark
sc = pyspark.SparkContext.getOrCreate()
```

```python
from faker import Faker
from datetime import date as dt
from random import random, choice
fake = Faker()
csv = []

for _ in range(1000):
    fecha = fake.date_between_dates(date_start=dt(2016,1,1),date_end=dt(2016,12,31)).strftime("%Y%m%d")
    driver_id = fake.pyint() % 50
    car_id = fake.pyint() % 50
    trip_id = fake.pyint()
    customer_id = fake.pyint()
    distance_traveled = fake.pyfloat(positive=True) % 1000.0
    csv.append((driver_id, car_id, trip_id, customer_id, fecha, distance_traveled))
```

```python
#(driver_id, car_id, trip_id, customer_id, date (YYYYMMDD), distance_traveled)
```

```python
rdd = sc.parallelize(csv)
```

```python
rdd.collect()
```

```python
rdd_2016_04 = rdd.filter(lambda x : x[4].startswith('201604'))
```

```python
rddDriverId = rdd_2016_04.map(lambda x : (x[0],(x[5], 1))).reduceByKey(lambda x,y : (x[0] + y[0], x[1] + y[1])).map(lambda x : (x[0], x[1][0]/x[1][1]))
```
